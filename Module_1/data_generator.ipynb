```python
import random

def generate_unique_name(indian_names, indian_surnames, full_names):
    while True:
        first = random.choice(indian_names)
        last = random.choice(indian_surnames)
        full_name = f"{first} {last}"

        if full_name not in full_names:
            full_names.add(full_name)   # assuming set
            return full_name

def generate_indian_mobile(mobile_numbers):
    while True:
        # Indian mobiles start with 6, 7, 8, or 9
        first_digit = random.choice(['6', '7', '8', '9'])
        remaining_digits = ''.join([str(random.randint(0, 9)) for _ in range(9)])
        mobile_no = f"{first_digit}{remaining_digits}"

        if mobile_no not in mobile_numbers:
            mobile_numbers.add(mobile_no)
            return mobile_no
```

```python
from faker import Faker
import random

fake = Faker()

def generate_chat(num_messages=5):
    users = [fake.name(), fake.name()]
    chat_log = []
    
    for _ in range(num_messages):
        message = {
            "sender": random.choice(users),
            "timestamp": fake.date_time_between(start_date="-1h", end_date="now").strftime("%H:%M:%S"),
            # Use .sentence() for short messages or .text() for longer ones
            "text": fake.sentence(nb_words=random.randint(5, 15))
        }
        chat_log.append(message)
    
    # Sort by timestamp to make the conversation chronological 
    return sorted(chat_log, key=lambda x: x['timestamp'])

# Display the chat
for msg in generate_chat(10):
    print(f"[{msg['timestamp']}] {msg['sender']}: {msg['text']}")
```

```python
import json

# Sample JSON-like data (list of dicts)

def json_to_sql(table_name, json_data):
    statements = []
    for entry in json_data:
        columns = ", ".join(entry.keys())
        
        # Format values: Add quotes to strings, handle None as NULL
        values = []
        for v in entry.values():
            if v is None:
                values.append("NULL")
            elif isinstance(v, str):
                # .replace("'", "''") escapes single quotes for SQL safety
                values.append(f"'{v.replace("'", "''")}'")
            else:
                values.append(str(v))
        
        value_str = ", ".join(values)
        statements.append(f"INSERT INTO {table_name} ({columns}) VALUES ({value_str});")
    
    return statements

# Usage
# sql_queries = json_to_sql("Members", generated_members)
# for query in sql_queries:
#     print(query)
```

```python
VehicleMap = {'Bike': 1,
    'Auto Rickshaw': 3,
    'Mini Cab': 4,
    'Sedan Cab': 4,
    'SUV Cab': 6,
    'Premium SUV': 7}
```

```python
import uuid
import random
import pandas as pd
from datetime import datetime, timedelta

members_df = pd.read_csv("Members.csv")
MEMBER_IDS = members_df['MemberID'].tolist()

IITGN = "IIT Gandhinagar (Palaj)"
LOCATIONS = ["Gift City", "Infocity", "Kudasan", "SVPI Airport", "Ahmedabad Railway Station", "SG Highway"]
PLATFORMS = ["Uber", "Ola", "InDrive", "Private"]

def generate_uuid():
    return str(uuid.uuid4())

def generate_synthetic_data(num_active=20, num_history=40):
    now = datetime.now()
    
    # --- Table 3: ActiveRides ---
    active_rides = []
    for _ in range(num_active):
        other_loc = random.choice(LOCATIONS)
        vehicle = random.choice(list(VehicleMap.keys()))
        max = VehicleMap[vehicle]
        pa = random.randint(1, max)
        active_rides.append({
            "RideID": generate_uuid(),
            "AdminID": random.choice(MEMBER_IDS),
            "AvailableSeats": max-pa,
            "PassengerCount": pa,
            "Source": random.choice([IITGN, other_loc]),
            "Destination": other_loc if random.choice([True, False]) else IITGN,
            "VehicleType": vehicle,
            "StartTime": (now + timedelta(hours=random.randint(2, 48))).strftime('%Y-%m-%d %H:%M:%S'),
            "EstimatedTime": random.randint(30, 90),
            "FemaleOnly": random.choice([True, False, False,False])
        })
    df_active = pd.DataFrame(active_rides)

    # --- Table 4: RidePassengerMap ---
    passenger_map = []
    for _, ride in df_active.iterrows():
        # Add 1-3 passengers to some rides
        if random.random() > 0.3:
            potential = [m for m in MEMBER_IDS if m != ride['AdminID']]
            selected = random.sample(potential, random.randint(1, 2))
            for p_id in selected:
                passenger_map.append({
                    "RideID": ride['RideID'],
                    "PassengerID": p_id,
                    "IsConfirmed": random.choice([True, False])
                })
    df_map = pd.DataFrame(passenger_map)

    # --- Table 10: RideHistory ---
    history = []
    for _ in range(num_history):
        other_loc = random.choice(LOCATIONS)
        history.append({
            "RideID": generate_uuid(), # Unique UUID for History record
            "AdminID": random.choice(MEMBER_IDS),
            "RideDate": (now - timedelta(days=random.randint(1, 20))).strftime('%Y-%m-%d'),
            "StartTime": f"{random.randint(6, 22):02d}:{random.choice(['00', '15', '30']):s}:00",
            "Source": random.choice([IITGN, other_loc]),
            "Destination": other_loc if random.choice([True, False]) else IITGN,
            "Platform": random.choice(PLATFORMS),
            "Price": random.randint(100, 600),
            "FemaleOnly": random.choice([True, False, False,False])
        })
    df_history = pd.DataFrame(history)

    # --- Table 8: RideFeedback (Linked to History UUIDs) ---
    feedback = []
    feedback_samples = df_history.sample(frac=0.5)
    for _, ride in feedback_samples.iterrows():
        feedback.append({
            "RideID": ride['RideID'], # Must exist in History
            "MemberID": random.choice(MEMBER_IDS),
            "FeedbackText": random.choice(["Smooth ride", "Driver was late", "Friendly co-passengers","Nice Car","AC not working","bad copassenger","Very great experience","Nice","Need to improve chat UI","Will recommned to friends","Awesome experience"]),
            "FeedbackCategory": random.choice(["Punctuality", "Safety", "Comfort"]),
            "SubmittedAt": datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        })
    df_feedback = pd.DataFrame(feedback)

    # --- Table 11: Cancellation (Linked to Active UUIDs) ---
    cancellations = []
    cancel_samples = df_active.sample(n=min(5, len(df_active)))
    for _, ride in cancel_samples.iterrows():
        cancellations.append({
            "RideID": ride['RideID'], # Must exist in ActiveRides
            "MemberID": random.choice(MEMBER_IDS),
            "CancellationReason": random.choice(["Plan Changed","Other private reasons","Bad co passengers"])
        })
    df_cancel = pd.DataFrame(cancellations)

    return df_active, df_map, df_history, df_feedback, df_cancel

# --- SQL Export Utility ---
def dataframe_to_sql(df, table_name):
    statements = []
    for _, row in df.iterrows():
        cols = ", ".join(row.index)
        vals = []
        for v in row.values:
            if isinstance(v, str): vals.append(f"'{v.replace("'", "''")}'")
            elif isinstance(v, bool): vals.append("TRUE" if v else "FALSE")
            else: vals.append(str(v))
        statements.append(f"INSERT INTO {table_name} ({cols}) VALUES ({', '.join(vals)});")
    return "\n".join(statements)

# Execution
df_active, df_map, df_history, df_feedback, df_cancel = generate_synthetic_data()
# print(df_active.head().to_markdown())

with open("seed_data_uuid.sql", "w") as f:
    f.write("-- Active Rides\n" + dataframe_to_sql(df_active, "ActiveRides") + "\n\n")
    f.write("-- Passenger Mapping\n" + dataframe_to_sql(df_map, "RidePassengerMap") + "\n\n")
    f.write("-- History\n" + dataframe_to_sql(df_history, "RideHistory") + "\n\n")
    f.write("-- Feedback\n" + dataframe_to_sql(df_feedback, "RideFeedback") + "\n\n")
    f.write("-- Cancellations\n" + dataframe_to_sql(df_cancel, "Cancellation") + "\n")

print("File 'seed_data.sql' created with UUID primary keys.")
```
