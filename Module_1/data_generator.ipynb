```python
import random

def generate_unique_name(indian_names, indian_surnames, full_names):
    while True:
        first = random.choice(indian_names)
        last = random.choice(indian_surnames)
        full_name = f"{first} {last}"

        if full_name not in full_names:
            full_names.add(full_name)   # assuming set
            return full_name

def generate_indian_mobile(mobile_numbers):
    while True:
        # Indian mobiles start with 6, 7, 8, or 9
        first_digit = random.choice(['6', '7', '8', '9'])
        remaining_digits = ''.join([str(random.randint(0, 9)) for _ in range(9)])
        mobile_no = f"{first_digit}{remaining_digits}"

        if mobile_no not in mobile_numbers:
            mobile_numbers.add(mobile_no)
            return mobile_no
```

```python
import random

# Top 20 Indian Names (10 Male, 10 Female)
indian_names = [
    "Aarav", "Vihaan", "Reyansh", "Advik", "Kiaan", 
    "Shaurya", "Ishaan", "Arjun", "Riaan", "Ayaan", 
    "Saanvi", "Aadhya", "Anaya", "Kiara", "Myra", 
    "Navya", "Inaaya", "Shanaya", "Vamika", "Diya"
]


# Top 20 Indian Surnames
indian_surnames = [
    "Singh", "Kumar", "Sharma", "Patel", "Gupta", 
    "Agarwal", "Yadav", "Verma", "Khan", "Mishra", 
    "Shah", "Reddy", "Jain", "Nair", "Das", 
    "Chaudhary", "Patil", "Mehta", "Rao", "Iyer"
]

iitgn_programmes = ["B.Tech.", "M.Tech.", "B.Tech.-M.Tech. Dual Degree", "PhD"]

iitgn_btech_branches = [
    "Electrical Engineering",
    "Computer Science and Engineering",
    "Mechanical Engineering",
    "Artificial Intelligence",
    "Chemical Engineering",
    "Civil Engineering",
    "Integrated Circuit Design & Technology",
    "Materials Engineering"
]

iitgn_mtech_branches = [
    "Artificial Intelligence",
    "Biological Engineering",
    "Chemical Engineering",
    "Civil Engineering",
    "Computer Science and Engineering",
    "Earth System Science",
    "Electrical Engineering",
    "Integrated Circuit Design and Technology",
    "Materials Engineering",
    "Mechanical Engineering"
]

iitgn_phd_branches = [
    "Biological Engineering", 
    "Chemical Engineering", 
    "Chemistry", 
    "Civil Engineering", 
    "Cognitive Science", 
    "Computer Science and Engineering", 
    "Artificial Intelligence", 
    "Earth Sciences", 
    "Archaeological Sciences", 
    "Electrical Engineering", 
    "Integrated Circuit Design and Technology", 
    "Humanities & Social Sciences", 
    "Materials Engineering", 
    "Mathematics", 
    "Mechanical Engineering", 
    "Design", 
    "Physics"
]

generated_members = []

# Generate random start
start_id = random.randint(10000000, 99999999)

full_names = set()
mobile_nos = set()

# Member Table
for i in range(30):
    # Randomly pick a name and a surname
    full_name = generate_unique_name(indian_names, indian_surnames, full_names)
    
    # Auto-increment MemberID
    member_id = start_id + i

    url = f"https://robohash.org/{member_id}.png?set=set4"

    first_name = full_name.split(" ")[0]
    
    gender = "M" if first_name in indian_names[:10] else "F"
    
    email_id = (".".join(full_name.split(" ")) + "@iitgn.ac.in").lower()

    programme = random.choice(iitgn_programmes)

        
    if programme == iitgn_programmes[1]:
        branch = random.choice(iitgn_mtech_branches)
        age = random.randint(22,27)
    elif programme == iitgn_programmes[3]:
        branch = random.choice(iitgn_phd_branches)
        age = random.randint(23, 35)
    else:
        branch = random.choice(iitgn_btech_branches)
        age = random.randint(18,23)

    batch_year = random.randint(2026,2031)
    
    ContactNumber = generate_indian_mobile(mobile_nos)
    

    generated_members.append({
        "MemberID": member_id,
        "FullName": full_name,
        "ProfileImageURL": url,
        "Programme": programme,
        "Branch": branch,
        "BatchYear": batch_year,
        "Email": email_id,
        "ContactNumber": ContactNumber,
        "Age": age,
        "Gender": gender
    })

# Preview the first 5
for member in generated_members[:5]:
    print(member)
```

```python
import json

# Sample JSON-like data (list of dicts)

def json_to_sql(table_name, json_data):
    statements = []
    for entry in json_data:
        columns = ", ".join(entry.keys())
        
        # Format values: Add quotes to strings, handle None as NULL
        values = []
        for v in entry.values():
            if v is None:
                values.append("NULL")
            elif isinstance(v, str):
                # .replace("'", "''") escapes single quotes for SQL safety
                values.append(f"'{v.replace("'", "''")}'")
            else:
                values.append(str(v))
        
        value_str = ", ".join(values)
        statements.append(f"INSERT INTO {table_name} ({columns}) VALUES ({value_str});")
    
    return statements

# Usage
sql_queries = json_to_sql("Members", generated_members)
for query in sql_queries:
    print(query)
```

```python
schema = """
CREATE TABLE IF NOT EXISTS Members (
    MemberID INT PRIMARY KEY,
    FullName VARCHAR(100) NOT NULL,
    ProfileImageURL VARCHAR(255) DEFAULT 'default_avatar.png',
    -- Points to a default image initially
    Programme VARCHAR(50) NOT NULL,
    Branch VARCHAR(50),
    BatchYear YEAR NOT NULL,            
    Email VARCHAR(100) NOT NULL UNIQUE,
    ContactNumber VARCHAR(15) NOT NULL UNIQUE,
    Age INT NULL,                      
    Gender CHAR(1) NULL                
);
"""
```

```python
def save_sql(sql):
    with open("members_dml.sql", "w", encoding="utf-8") as f:
        # Optional: Add transaction headers for safety and speed
        f.write("BEGIN TRANSACTION;\n\n")
        f.write(schema + "\n")
        # Write each SQL statement on a new line
        for statement in sql_queries:
            f.write(statement + "\n")
        
        f.write("\nCOMMIT;")

    print("Successfully generated members_dml.sql")
```
